```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)

file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame

purchased_data = pd.read_csv(file_to_load)
purchased_data.head()
```

```python
#Display the total number of players

counted_data = purchase_data.groupby("SN")["Gender"].count()
counted_data_2 = counted_data.count()
df2 = pd.DataFrame({"total players":[counted_data_2]})
df2

```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

ID = purchase_data.groupby("Item ID")["Item Name"].count()
ID_2 = ID.count()
avg = round(purchase_data["Price"].mean(),2)
num_pur = purchase_data["Purchase ID"].count()
reve = purchase_data["Price"].sum()
df3 = pd.DataFrame({"Number of Unique Items":[ID_2], 
                    "Average Price": "$" +  str(avg), 
                    "Number of Purchases":[num_pur], 
                    "Total Revenue":"$" + str(reve)})
df3
```

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

counted_data_3 = purchase_data.groupby("Gender")["SN"].count()

avg_2 = purchase_data.groupby("Gender")["Price"]
avg_3 = round(avg_2.mean(),2)

total = purchase_data.groupby("Gender")["Price"]
total_2 = round(total.sum(),2)

avg_total = round(total.sum()/avg_3,2)

counted_data_5 = pd.DataFrame({"Purchase Count": counted_data_3,
                               "Average Purchase Price": avg_3,
                              "Total Purchase Value": total_2 })
counted_data_5
```

```python
#Establish bins for ages
#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table

#age_1 = purchase_data.groupby("Age")["SN"]
#age_2 = age_1.count([0-10],[10-14])
#age_2

#counted_data_6 = pd.DataFrame({"col 1":["0-10","10-14","15-19","20-24","25-29","30-34","35-39","40+"] })
#counted_data_6.sort_values(by ="col 1", ascending=True)
#counted_data_6=[]
bins = [0,10,14,19,24,29,34,39,50]
group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchased_data["Total Count"] = pd.cut(purchased_data["Age"], bins, labels=group_names)
#counted_data_6 = purchased_data.groupby("Total Count")["Age"].count()
#counted_data_7 = pd.DataFrame(counted_data_6).sort_values(by="Total Count", ascending = False)
#counted_data_8 = counted_data_7.groupby("Total Count").count()
#counted_data_8
purchased_data_2 = purchased_data[["Item ID","Total Count"]]
purchased_data_3 = purchased_data_2.groupby("Total Count").count()
purchased_data_3["Percentage of Players"] = round((purchased_data_3["Item ID"]/purchased_data_3["Item ID"].sum())*100,2)
purchased_data_4 = purchased_data_3.rename(columns={"Item ID":"Total Count"})
purchased_data_4
```

```python
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

avg_4 = purchased_data.groupby("Total Count")["Price"]
avg_5 = round(avg_4.mean(),2)
purchased_data_4["Average Purchase Price"] = avg_5
purchased_data_5 = purchased_data_4[["Total Count", "Average Purchase Price"]]
purchased_data_5["Total Purchase Value"] = purchased_data_5["Total Count"]*purchased_data_5["Average Purchase Price"]
purchased_data_5["Avg Total Purchase per Person"] = round((purchased_data_5["Total Purchase Value"]/purchased_data_5["Total Purchase Value"].sum())*100,2)
purchased_data_5
```
